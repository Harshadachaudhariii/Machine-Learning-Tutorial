```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, centers= 3, n_features=2, random_state=1)
```

```python
X
```

```python
y    #in usl y not use
```

```python
plt.scatter(X[:, 0],X[:,1], c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans
```

```python
kmeans.fit(X_train)
```

```python
y_label = kmeans.predict(X_train)
```

```python
plt.scatter(X_train[:, 0],X_train[:,1], c=y_label)
```

```python
# how better the clustering model is

from sklearn.metrics import silhouette_score
silhouette_score(X_train,kmeans.labels_)
```

```python
kmeans.labels_
```

```python
kmeans.inertia_   #for k=3, you have wcss distance
```

```python
# to select k(centroid) --> elbow method
wcss = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_ )
```

```python
wcss
```

```python
# plot of elbow curve
plt.plot(range(1,10),wcss)
plt.xticks(range(1,10))
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
# based on the above plot the optimal no of k cluster is 3
# k=3 3 where elbow method will be constant 
```
