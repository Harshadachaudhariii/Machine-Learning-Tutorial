```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.keys()  
```

```python
# problem statement --> to predict disease progression one year after baseline 
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
data = pd.DataFrame(diabetes.data, columns= diabetes.feature_names)
```

```python
data
```

```python
data['target'] = diabetes.target
# merge the data
```

```python
data
```

```python
# data preparation  -> here already the data ready
```

```python
# divide the data x and y 
```

```python
# simple linear regression 
X = data[['bmi']]
y = data['target']
```

```python
X
```

```python
y
```

```python
# train-test splits 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y, test_size= 0.20, random_state=111)   #random_state lock randomness
X_train                                                 #train_size = 0.80
```

```python
# scaling is done in this datasets
# model training
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
# now train the model
model.fit(X_train, y_train)
```

```python
# y=mx+c m value, beta1
model.coef_
```

```python
# c value, betanot
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# visualize the result
plt.scatter(X_test, y_test, color= 'black', label= 'Actual Data')
plt.plot(X_test, y_pred, color ='red', linewidth=3, label= 'linear regression line')
plt.xlabel("BMI")
plt.ylabel("One year progression-target ")
plt.title("My first linear regression model ")
plt.legend()
plt.show()
```

```python
# equation of line --> y =967.57 * x +150
```

```python
# evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
# RMSE
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
score=r2_score(y_test, y_pred)
score
```

```python
# adj r2
1- (1-score)*(len(y_test)-1) /  (len(y_test) - X_test.shape[1]-1)
```
